## QUESTION 1

**Answer:**  
This follows a **Binomial distribution** (n = 30, p = 1/6, k = 5).  
**Probability = 0.1921** (or 19.21%)

## QUESTION 2  
Generate 100 values in Excel using the continuous uniform distribution RAND() and plot a histogram. Describe the shape of the distribution.

**Answer:**  
The histogram of a continuous uniform distribution (values between 0 and 1) will be approximately **flat / rectangular** in shape.  
All bars will have roughly equal height because every value between 0 and 1 has equal probability.

**EXCEL VERSION (Exact Steps):**  
1. Open Excel  
2. In cell **A1** type `=RAND()`  
3. Drag down to **A100** (100 random numbers between 0 and 1)  
4. Select range A1:A100  
5. Go to **Insert → Charts → Histogram** (or use Data Analysis Toolpak → Histogram)  
6. The chart will show almost equal height bars → **flat rectangular shape**

## QUESTION 3  
A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?

**Answer:**  
45 = mean – 1 SD  
55 = mean + 1 SD  
**Approximately 68.27%** of the values lie between 45 and 55 (Empirical Rule / 68-95-99.7 rule).

## QUESTION 4  
What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

**Answer:**  
**Standardization (z-score)** converts data into a standard normal distribution with mean = 0 and standard deviation = 1.  
**Formula:**  
z = (x – μ) / σ  

**Why important?**  
- Allows comparison of values from different scales/units  
- Helps detect outliers  
- Required for many machine learning algorithms and statistical tests  

It transforms any dataset to have mean 0 and SD 1.

## QUESTION 5  
What is Kurtosis and their type?

**Answer:**  
**Kurtosis** measures the "tailedness" or peakedness of a distribution compared to a normal distribution.  
**Types:**  
- **Leptokurtic** (Kurtosis > 3): Heavy tails, sharp peak (more outliers)  
- **Mesokurtic** (Kurtosis = 3): Normal distribution shape  
- **Platykurtic** (Kurtosis < 3): Light tails, flat peak (fewer outliers)

## QUESTION 6  
Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

**Answer:**  
A fair die has 6 equally likely outcomes (1 through 6).  
In a **discrete uniform distribution**, every outcome has exactly the same probability (1/6).  
This matches the definition of uniform distribution where all possible values are equally likely.

## QUESTION 7  
Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5

**Answer:**  
This is Binomial (n = 15, p = 0.5).  
**Probability = 0.5** (exactly 50%)  
In Excel: `=1 - BINOM.DIST(7,15,0.5,TRUE)`

** EXCEL VERSION (Exact Steps + Formula):**  
1. Open Excel  
2. In any cell (e.g. **C3**) type this formula:  
   `=1 - BINOM.DIST(7,15,0.5,TRUE)`  
3. Press Enter → Result will be exactly **0.5**

## QUESTION 8  
How does log transformation help in stabilizing variance and making data more normally distributed?

**Answer:**  
Log transformation:  
- **Stabilizes variance** (reduces heteroscedasticity)  
- **Compresses right-skewed data** (common in income, sales, etc.)  
- Makes the distribution more symmetric and closer to **normal**  
This helps meet assumptions of many statistical models and improves model performance.
